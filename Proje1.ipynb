```python
#Dersler={1:"Matematik",2:"Geometri",3:"Fen Bilgisi",4:"Kimya",5:"Fizik"}
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def Not_Araligi(puan):
  Harf_Notu=""
  Gecti_Mi=""
  if puan<45 and puan>=0:
    Harf_Notu="E"
    Gecti_Mi="Kalmıştır"
  elif puan<55 and puan>=45:
    Harf_Notu="D"
    Gecti_Mi="Geçmiştir"
  elif puan<70 and puan>=55:
    Harf_Notu="C"
    Gecti_Mi="Geçmiştir"
  elif puan<85 and puan>=70:
    Harf_Notu="B"
    Gecti_Mi="Geçmiştir"
  elif puan<95 and puan>=85:
    Harf_Notu="A"
    Gecti_Mi="Geçmiştir"
  elif puan<100 and puan>=95:
    Harf_Notu="A+"
    Gecti_Mi="Geçmiştir"
  return Harf_Notu,Gecti_Mi
```

```python
gecti_kaldi_df=pd.DataFrame(columns=["ID","Ad","Soyad","Ders","Puan","Harf Notu","Durumu"])
Cebir1=dict()
def Kayit(id,ad,soyad,ders,puan):
  Harf_Notu,Gecti_Mi=Not_Araligi(puan)
  Cebir1[id]=[ad,soyad,ders,puan]
  gecti_kaldi_df.loc[-1]=[id,ad,soyad,ders,puan,Harf_Notu,Gecti_Mi]
  gecti_kaldi_df.reset_index(drop=True,inplace=True)
  print(ad,soyad,Harf_Notu,"ile",Gecti_Mi,".")
  return gecti_kaldi_df

def Ogrenci_Kaydi():
    a=str(input("Öğrencinin numarasını giriniz: "))
    b=str(input("Öğrencinin adını giriniz: "))
    c=str(input("Öğrencinin soyadını giriniz: "))
    d=str(input("Eklenecek dersi giriniz: "))
    e=float(input("Dersin Notunu Giriniz: "))
    Kayit(a,b,c,d,e)
    print("Kaydınız başarıyla tamamlanmıştır.")
    #with open("notlar.txt","a",encoding="utf-8") as file:
```

```python
Kayit(11,"Ömer","Tekin","Fizik",75)
Kayit(12,"Tamer","Tekin","Matematik",65)
Kayit(23,"Büşra","Köksal","Matematik",70)
Kayit(34,"Zeynep","Akkoç","Fizik",90)
Kayit(54,"Tuana","Kurşun","Kimya",65)
Kayit(10,"Zeynep","Aslan","Türkçe",65)
Kayit(37,"Beyza","Köksal","Kimya",65)
```

```python
while True:
  print("Yapılacak işlemler \n 1: Yeni öğrenci kaydı \n 2: Öğrenci not döküm tablosunu görüntüleme \n Sistemden çıkış")
  x = input("Yapılacak işlemi seçiniz: ")
  if (int(x)==1):
    Ogrenci_Kaydi()
  elif (int(x)==2):
    print(gecti_kaldi_df)
  elif (int(x)==3):
    print("Sistemden çıktınız.")
    break
  else:
    print("Hatalı işlem!!!")
```
