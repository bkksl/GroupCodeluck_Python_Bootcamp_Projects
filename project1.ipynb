```python
#Dersler={1:"Matematik",2:"Geometri",3:"Fen Bilgisi",4:"Kimya",5:"Fizik"}
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def Not_Araligi(puan):
  Harf_Notu=""
  Gecti_Mi=""
  if puan<45 and puan>=0:
    Harf_Notu="E"
    Gecti_Mi="Kalmıştır"
  elif puan<55 and puan>=45:
    Harf_Notu="D"
    Gecti_Mi="Geçmiştir"
  elif puan<70 and puan>=55:
    Harf_Notu="C"
    Gecti_Mi="Geçmiştir"
  elif puan<85 and puan>=70:
    Harf_Notu="B"
    Gecti_Mi="Geçmiştir"
  elif puan<95 and puan>=85:
    Harf_Notu="A"
    Gecti_Mi="Geçmiştir"
  elif puan<100 and puan>=95:
    Harf_Notu="A+"
    Gecti_Mi="Geçmiştir"
  return Harf_Notu,Gecti_Mi

gecti_kaldi_df=pd.DataFrame(columns=["ID","Ad","Soyad","Puan","Harf Notu","Durumu"])
Cebir1=dict()
def Cebir(id,ad,soyad,puan):
  Harf_Notu,Gecti_Mi=Not_Araligi(puan)
  Cebir1[id]=[ad,soyad,puan]
  gecti_kaldi_df.loc[-1]=[id,ad,soyad,puan,Harf_Notu,Gecti_Mi]
  gecti_kaldi_df.reset_index(drop=True,inplace=True)
  print(ad,soyad,Harf_Notu,"ile",Gecti_Mi,".")
  return gecti_kaldi_df

```

```python
Cebir(11,"ömer","tekin",75)
Cebir(12,"tamer","tekin",65)
```
